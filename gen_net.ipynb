```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import math

from torch import optim
from tqdm import tqdm
```

```python
xs = torch.tensor([[0.,0.],[1.,0.],[0.,1.],[1.,1.]])

# and
# ys = torch.tensor([[0],[0],[0],[1]]).float()
# nand
# ys = torch.tensor([[1],[1],[1],[0]]).float()
# xor
# ys = torch.tensor([[0],[1],[1],[0]]).float()
# or
# ys = torch.tensor([[0],[1],[1],[1]]).float()
# nor
# ys = torch.tensor([[1],[0],[0],[0]]).float()
# ~xor
ys = torch.tensor([[1],[0],[0],[1]]).float()
```

```python
model = nn.Sequential(nn.Linear(2, 16), nn.ReLU(), nn.Linear(16, 1), nn.Sigmoid())
```

```python
EPOCHS = 5_000
opt = optim.Adam(model.parameters(), lr=1e-3)
lossf = nn.BCELoss()

for _ in tqdm(range(EPOCHS)):
    opt.zero_grad()
    loss = lossf(model(xs), ys)
    
    print(f"{loss.item()=}")

    loss.backward()
    opt.step()
```

```python
import itertools
out_text = ""
out_text += "# out_vec is also used as a doublebuffer\n"
out_text += "function runnet(vec, out_vec,  mat, bias)\n"
out_text += "{\n"
out_text += "\t# init variable\n"
out_text += "\tout_vec[0]=-1\n"

vec_size = None
for mod in model:
    if type(mod) is nn.Linear:
        for i, row in enumerate(mod.weight):
            for j, elem in enumerate(row):
                out_text += f"\tmat[{i},{j}]={elem.item():.5f}\n"
        
        n, m = mod.weight.shape
        vec_size = n
        out_text += f"\tmat_vec_mul(mat, vec, {n}, {m}, out_vec)\n"
        
        for i, elem in enumerate(mod.bias):
            out_text += f"\tbias[{i}]={elem.item():.5f}\n"
        
        out_text += f"\tvec_add(out_vec, bias, vec, {n})\n"
    elif type(mod) is nn.ReLU:
        out_text += f"\trelu(vec, out_vec, {vec_size})\n"
        out_text += f"\tcopy_vec(out_vec, vec, {vec_size})\n"
    elif type(mod) is nn.Sigmoid:
        out_text += f"\tsigmoid(vec, out_vec, {vec_size})\n"
        out_text += f"\tcopy_vec(out_vec, vec, {vec_size})\n"
        
out_text += "}\n"
```
